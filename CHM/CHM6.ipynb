```python
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sc
import sympy as sp
```

```python
h = 0.025
x = [round(0 + h*i,3) for i in range(41)]
print(np.array(x))
```

```python
N = 5
f = sp.Function('f')
t = sp.Symbol('\u03C4')
f = sp.sin(math.pi * t) * (12 + 25*t)**(1/2)
f
```

```python
y = [round(f.subs(t,i),8) for i in x]

print(np.array(y))
```

```python
a = []
for i in range(40):
    a.append(y[i])

print(np.array(a))
```

```python
g = []
for i in range(1,41):
    g.append((y[i] - y[i-1])/h)

print(np.array(g))
```

```python
ff = ["ff"+ str(i) for i in range(41)]
cn = [sp.Symbol("c"+ str(i)) for i in range(1,41)]

for i in range(38):
    ff[i+1] = sp.Eq(h*cn[i] + 4*h*cn[i+1] + h*cn[i+2], 3*(g[i+1] - g[i]))
    
ff[0] = sp.Eq(cn[0],0)
ff[39] = sp.Eq(cn[39],0)

for i in range(40):
    print(ff[i])
```

```python
cc=sp.solve(ff)
c =[cc.get(i) for i in cn]

print(np.array(c))
```

```python
k=39
b = []
for i in range(39):
    b.append(g[i] - (h/3)*(2*c[i] + c[i+1]))
b.append(b[k-1] + h*(c[k-1]+c[k]))

print(np.array(b))
```

```python
d = []
for i in range(39):
    d.append((1/(3*h))*(c[i+1]-c[i]))
d.append((1/(h**3))*(y[k] - a[k] - h*b[k] - c[k]*(h**2)))

print(np.array(d))
```

```python
P = ["P"+ str(i) for i in range(41)]

for i in range(40):
    P[i] = a[i] + b[i]*(t-x[i]) + c[i]*((t-x[i])**2) + d[i]*((t-x[i])**3)
    
for i in range(40):
    print(P[i])
```

```python
yP = []
for i in range(40):
    yP.append(P[i].subs(t,x[i]))

print(np.array(yP))
```

```python
xx = []
for i in range(40):
    xx.append(x[i])
plt.figure(figsize=(11,7))
plt.plot(x,y,color = "orange", linewidth=7.0)
plt.plot(xx,yP,color = "black")
```
