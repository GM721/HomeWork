```python
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sc
import sympy as sp
```

```python
x = [round(-1 + 0.1*i,1) for i in range(21)]
for i in x: print(i, end=', ')
x
```

```python
N = 5
f = sp.Function('f')
t = sp.Symbol('\u03C4') 
f = (10 + N) / (1 + (25 + 0.3*N)*(1 + 0.1)*t**2)
f
```

```python
y = [f.subs(t,i) for i in x]
for i in y: print(i)
```

```python
def A(t,x):
    g = sp.Function('g')
    g = 1
    for j in x:
        if t != j:
            g = g * (t - j)
    return g
```

```python
L = sp.Function('L')
L = 0
for i in range(len(x)):
    L = L +(y[i]/A(x[i],x))*sp.simplify(A(t,x)/(t - x[i]))
L = sp.simplify(L)
L
```

```python
xlin = np.linspace(-1,1,41)
xlin2 = np.linspace(-1,1,41)
L2 = [L.subs(t, i) for i in xlin]
f2 = [f.subs(t, i) for i in xlin2]
plt.figure(figsize=(11,7))
plt.plot(xlin,L2,color = "black")
plt.plot(xlin2,f2,color = "blue")
```

```python
xlin3=[]
for i in xlin:
    if abs(i) < 0.75:
        xlin3.append(i)
xlin2 = np.linspace(-1,1,41)
L2 = [L.subs(t, i) for i in xlin3]
f2 = [f.subs(t, i) for i in xlin2]
plt.figure(figsize=(11,7))
plt.plot(xlin3,L2,color = "black")
plt.plot(xlin2,f2,color = "blue")
```

```python
yy = [f.subs(t,i) for i in xx]
for i in yy: print(i)
```
